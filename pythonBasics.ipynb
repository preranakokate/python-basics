6.Why are lists mutable, but tuples are immutable?

- Lists are designed for frequent modification (add, remove, change items).

- Tuples are designed for fixed data that should not change.

- Tuples are faster and more memory-efficient.

- Immutability makes tuples safe (can be used as dictionary keys).

In short:
Lists = flexible
Tuples = fixed & safe

```python
# 6. Write a function to demonstrate the behavior of mutable and immutable arguments.
# Function to modify arguments
def modify_data(my_list, my_number):
    my_list.append(100)      # Mutable object change
    my_number += 10           # Immutable object change
    print("Inside function - list:", my_list)
    print("Inside function - number:", my_number)

# Mutable and immutable variables
numbers_list = [1, 2, 3]    # Mutable
num = 50                     # Immutable

# Call the function
modify_data(numbers_list, num)

# Check values after function call
print("Outside function - list:", numbers_list)  # List is changed
print("Outside function - number:", num)        # Number remains unchanged
```

```python
# 10. Write code to demonstrate type casting with list elements.
# List with string numbers
my_list = ["10", "20", "30"]

# Type casting list elements to integers
int_list = []
for item in my_list:
    int_list.append(int(item))

print("Original list (strings):", my_list)
print("List after type casting to int:", int_list)

# Type casting list elements to float
float_list = []
for item in my_list:
    float_list.append(float(item))

print("List after type casting to float:", float_list)
```

```python
# 14.Write a program to reverse a string using a while loop.
text = input("Enter a string: ")

# Reverse using while loop
rev = ""
i = len(text) - 1

while i >= 0:
    rev += text[i]
    i -= 1

print("Reversed string:", rev)
```

```python
# 15.Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is:", factorial)

```
